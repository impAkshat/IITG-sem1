```python
import numpy as np

# Define the 15 valid states as specified in the textbook equation (44.11)
# We map them to indices 0 through 14 for the NumPy array
states = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, "EXIT"]
state_to_idx = {s: i for i, s in enumerate(states)}

P = np.zeros((15, 15))

def add_transition(s_from, s_to, prob):
    idx_from = state_to_idx[s_from]
    idx_to = state_to_idx[s_to]
    P[idx_from, idx_to] += prob  # Use += to accumulate bounce probabilities


add_transition(1, 1, 0.50)  # Bounces right wall, bottom wall
add_transition(1, 2, 0.25)  # Left
add_transition(1, 8, 0.25)  # Up

# State 2
add_transition(2, 2, 0.25)  # Bounces bottom wall
add_transition(2, 1, 0.25)  # Right
add_transition(2, 3, 0.25)  # Left
add_transition(2, 7, 0.25)  # Up

# State 3
add_transition(3, 3, 0.50)  # Bounces bottom wall, Obstacle #4
add_transition(3, 2, 0.25)  # Right
add_transition(3, 6, 0.25)  # Up

# State 5
add_transition(5, 5, 0.50)  # Bounces left wall, Obstacle #4
add_transition(5, 6, 0.25)  # Right
add_transition(5, 12, 0.25) # Up

# State 6
add_transition(6, 6, 0.25)  # Bounces Obstacle #11
add_transition(6, 3, 0.25)  # Down
add_transition(6, 5, 0.25)  # Left
add_transition(6, 7, 0.25)  # Right

# State 7 (Calculated in textbook example)
add_transition(7, 2, 0.25)  # Down
add_transition(7, 6, 0.25)  # Left
add_transition(7, 8, 0.25)  # Right
add_transition(7, 10, 0.25) # Up

# State 9
add_transition(9, 9, 0.25)  # Bounces right wall
add_transition(9, 8, 0.25)  # Down
add_transition(9, 10, 0.25) # Left
add_transition(9, 16, 0.25) # Up

# State 10
add_transition(10, 10, 0.25) # Bounces Obstacle #11
add_transition(10, 7, 0.25)  # Down
add_transition(10, 9, 0.25)  # Right
add_transition(10, 15, 0.25) # Up

# State 12
add_transition(12, 12, 0.50) # Bounces left wall, Obstacle #11
add_transition(12, 5, 0.25)  # Down
add_transition(12, 13, 0.25) # Up

# State 13
add_transition(13, 13, 0.50) # Bounces top wall, left wall
add_transition(13, 12, 0.25) # Down
add_transition(13, 14, 0.25) # Right

# State 14
add_transition(14, 14, 0.50) # Bounces top wall, Obstacle #11
add_transition(14, 13, 0.25) # Left
add_transition(14, 15, 0.25) # Right

# State 15
add_transition(15, 15, 0.25) # Bounces top wall
add_transition(15, 10, 0.25) # Down
add_transition(15, 14, 0.25) # Left
add_transition(15, 16, 0.25) # Right

# ==========================================================
# 2. Terminal States and EXIT
# Equations (44.18a), (44.18b), and (44.19)
# ==========================================================

# State 8 (Danger) -> Action is STOP
add_transition(8, "EXIT", 1.0)

# State 16 (Star) -> Action is STOP
add_transition(16, "EXIT", 1.0)

# EXIT State -> Absorbing state
add_transition("EXIT", "EXIT", 1.0)

# ==========================================================
# Validation Check
# ==========================================================
# Assert that every row sums to 1.0 to ensure it is a valid Markov Matrix
assert np.allclose(P.sum(axis=1), 1.0), "Error: Not all rows sum to 1.0!"

# Print the specific example from the textbook image
print(f"Probability from 7 to 10 (Textbook example): {P[state_to_idx[7], state_to_idx[10]]}")
print(f"Probability from 8 to EXIT: {P[state_to_idx[8], state_to_idx['EXIT']]}")
```
